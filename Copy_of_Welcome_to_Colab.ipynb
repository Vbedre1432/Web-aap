```
register_php_content = """
<?php
require 'vendor/autoload.php'; // Assuming you're using Composer for the MongoDB driver
require 'config.php'; // Database configuration

use MongoDB\\Client;
use MongoDB\\Driver\\Exception\\BulkWriteException;

header('Content-Type: application/json');

if ($_SERVER['REQUEST_METHOD'] === 'POST') {
    // 1. Get input data
    $username = $_POST['username'] ?? '';
    $email = $_POST['email'] ?? '';
    $password = $_POST['password'] ?? '';

    // 2. Validate input data
    if (empty($username) || empty($email) || empty($password)) {
        echo json_encode(['success' => false, 'message' => 'Please fill in all fields.']);
        exit;
    }

    if (!filter_var($email, FILTER_VALIDATE_EMAIL)) {
        echo json_encode(['success' => false, 'message' => 'Invalid email format.']);
        exit;
    }

    // 3. Hash the password
    $hashedPassword = password_hash($password, PASSWORD_DEFAULT);

    // Connect to MongoDB and insert user
    try {
        $client = new Client(MONGO_URI);
        $db = $client->selectDatabase(MONGO_DB);
        $usersCollection = $db->users;

        $newUser = [
            'username' => $username,
            'email' => $email,
            'password' => $hashedPassword,
            'registration_date' => new MongoDB\\BSON\\UTCDateTime(),
            'role' => 'student' // Default role
        ];

        $insertResult = $usersCollection->insertOne($newUser);

        if ($insertResult->getInsertedCount() === 1) {
            echo json_encode(['success' => true, 'message' => 'Registration successful.']);
        } else {
            echo json_encode(['success' => false, 'message' => 'Registration failed.']);
        }

    } catch (BulkWriteException $e) {
        $errors = $e->getWriteResult()->getWriteErrors();
        $isDuplicate = false;
        foreach ($errors as $error) {
            // Check for duplicate key error code (11000)
            if ($error->getCode() === 11000) {
                $isDuplicate = true;
                break;
            }
        }
        if ($isDuplicate) {
            echo json_encode(['success' => false, 'message' => 'Username or email already exists.']);
        } else {
            echo json_encode(['success' => false, 'message' => 'Database error: ' . $e->getMessage()]);
        }
    } catch (Exception $e) {
        echo json_encode(['success' => false, 'message' => 'An error occurred: ' . $e->getMessage()]);
    }
} else {
    echo json_encode(['success' => false, 'message' => 'Invalid request method.']);
}
?>
"""

login_php_content = """
<?php
require 'vendor/autoload.php'; // Assuming you're using Composer
require 'config.php'; // Database configuration

use MongoDB\\Client;

header('Content-Type: application/json');
session_start(); // Start the PHP session

if ($_SERVER['REQUEST_METHOD'] === 'POST') {
    // 4. Get input data
    $identifier = $_POST['identifier'] ?? ''; // Can be username or email
    $password = $_POST['password'] ?? '';

    if (empty($identifier) || empty($password)) {
        echo json_encode(['success' => false, 'message' => 'Please enter username/email and password.']);
        exit;
    }

    // 5. Connect to MongoDB and retrieve user
    try {
        $client = new Client(MONGO_URI);
        $db = $client->selectDatabase(MONGO_DB);
        $usersCollection = $db->users;

        // Try to find user by username or email
        $user = $usersCollection->findOne([
            '$or' => [
                ['username' => $identifier],
                ['email' => $identifier]
            ]
        ]);

        if ($user) {
            // 6. Verify password
            if (password_verify($password, $user['password'])) {
                // 7. Start session and store user info
                $_SESSION['user_id'] = (string) $user['_id']; // Store MongoDB ID as string
                $_SESSION['username'] = $user['username'];
                $_SESSION['role'] = $user['role'];

                echo json_encode(['success' => true, 'message' => 'Login successful.']);
            } else {
                echo json_encode(['success' => false, 'message' => 'Invalid username/email or password.']);
            }
        } else {
            echo json_encode(['success' => false, 'message' => 'Invalid username/email or password.']);
        }

    } catch (Exception $e) {
        echo json_encode(['success' => false, 'message' => 'An error occurred: ' . $e->getMessage()]);
    }
} else {
    echo json_encode(['success' => false, 'message' => 'Invalid request method.']);
}
?>
"""

session_check_php_content = """
<?php
// This is a conceptual example of how session checking would work
// You would include this at the top of pages requiring authentication

session_start();

// 8. Check if the user is logged in
function check_session() {
    if (!isset($_SESSION['user_id'])) {
        // User is not logged in, redirect to login page
        header('Location: login.html'); // Or your login page path
        exit;
    }
    // User is logged in, continue loading the page
}

// Call this function at the beginning of protected pages
// check_session();

// Example of how to access session data:
// $loggedInUserId = $_SESSION['user_id'];
// $loggedInUsername = $_SESSION['username'];
// $loggedInUserRole = $_SESSION['role'];
?>
"""

logout_php_content = """
<?php
// 9. Logout script
session_start(); // Start the session

// Unset all session variables
$_SESSION = array();

// Destroy the session
session_destroy();

// Redirect to the login page
header('Location: login.html'); // Or your login page path
exit;
?>
"""

config_php_content = """
<?php
// Database configuration
define('MONGO_URI', 'mongodb://localhost:27017/'); // Replace with your MongoDB connection string
define('MONGO_DB', 'learning_platform_db'); // Replace with your database name
?>
"""

print("---register.php---")
print(register_php_content)
print("\n---login.php---")
print(login_php_content)
print("\n---session_check_logic.php (conceptual)---")
print(session_check_php_content)
print("\n---logout.php---")
print(logout_php_content)
print("\n---config.php---")
print(config_php_content)
```
